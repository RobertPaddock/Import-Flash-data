```python
#This cell loops over all the files. It generates the fixed resolution buffer, and then saves this to a csv file. 

#Loop over number of files
filenums = range(0,700)
for i in list(filenums):

    #Get filename in correct format
    if i<10:
        string = "lasslab_hdf5_plt_cnt_000" + str(i)
        
    elif 9<i<100:
        string = "lasslab_hdf5_plt_cnt_00" + str(i)
        
    elif 99<i:
        string = "lasslab_hdf5_plt_cnt_0" + str(i)
        
    namestring = "Pressure" + str(i) + ".csv"

    #Load the file
    ds = yt.load(string)
    
    #Generate the fixed resolution buffer. Load the density from this.
    slice = ds.slice("z", 0)
    width = (0.056,  "cm")  # the appropriate width and center needs to be determined. Can see what the buffer looks like using the yt_plots script.
    res = [500, 500]  
    frb = slice.to_frb(width, res, center=[0.028,0.028, 0])
    density = frb["pres"].d
    
    #Write to a csv file
    with open(namestring,"w+") as my_csv:
        csvWriter = csv.writer(my_csv,delimiter=',')
        csvWriter.writerows(density)
```

```python
#This cell loops again, and saves all the time steps to a csv file as well.

List = []

#Loop over number of files
filenums = range(0,523)
for i in list(filenums):

    #Get filename in correct format
    if i<10:
        string = "lasslab_hdf5_plt_cnt_000" + str(i)
        
    elif 9<i<100:
        string = "lasslab_hdf5_plt_cnt_00" + str(i)
        
    elif 99<i:
        string = "lasslab_hdf5_plt_cnt_0" + str(i)
        
    namestring = "File" + str(i) + ".csv"

    #Load the file
    ds = yt.load(string)
   

    Current_Time = float(ds.current_time.to('s'))
    List.append(Current_Time)


#Write to a csv file
with open("timearray.csv","w+") as my_csv:
    csvWriter = csv.writer(my_csv)
    csvWriter.writerows(map(lambda x: [x], List))
    
```
